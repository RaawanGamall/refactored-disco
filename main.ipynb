```python
# ------------------------
# Text Cleaning
# ------------------------
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# ------------------------
# Load Data
# ------------------------
df = pd.read_csv("clean_train.csv", encoding='ISO-8859-1')
df.dropna(subset=['text', 'sentiment', 'selected_text'], inplace=True)
df['clean_text'] = df['text'].apply(clean_text)
df['clean_selected_text'] = df['selected_text'].apply(clean_text)
```

```python
# ------------------------
# TF-IDF + Logistic Regression
# ------------------------
vectorizer = TfidfVectorizer(ngram_range=(1,2), max_features=5000)
X = vectorizer.fit_transform(df['clean_text']).toarray()
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['sentiment'])
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

def one_hot(y, num_classes):
    return np.eye(num_classes)[y]

y_train_oh = one_hot(y_train, 3)
y_test_oh = one_hot(y_test, 3)

def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def cross_entropy_loss(y_true, y_pred):
    return -np.mean(np.sum(y_true * np.log(y_pred + 1e-9), axis=1))

def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

def train_logistic_regression(X, y, lr=0.1, epochs=1000):
    m, n = X.shape
    num_classes = y.shape[1]
    W = np.zeros((n, num_classes))
    b = np.zeros((1, num_classes))

    for epoch in range(epochs):
        logits = np.dot(X, W) + b
        probs = softmax(logits)
        loss = cross_entropy_loss(y, probs)

        grad_W = np.dot(X.T, (probs - y)) / m
        grad_b = np.mean(probs - y, axis=0, keepdims=True)

        W -= lr * grad_W
        b -= lr * grad_b

        if epoch % 100 == 0:
            print(f"Epoch {epoch} - Loss: {loss:.4f}")

    return W, b

W, b = train_logistic_regression(X_train, y_train_oh)

def predict(X, W, b):
    logits = np.dot(X, W) + b
    probs = softmax(logits)
    return np.argmax(probs, axis=1)

y_pred = predict(X_test, W, b)

print("\n[Logistic Regression Metrics]")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
print("Precision:", precision_score(y_test, y_pred, average='weighted'))
print("Recall:", recall_score(y_test, y_pred, average='weighted'))

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Logistic Regression Confusion Matrix")
plt.show()
```

```python

# Define input layer with fixed sequence length
input_cnn = Input(shape=(max_len,), name="input_layer")

# Embedding layer maps each word to a 128-dim vector
embedding = Embedding(input_dim=max_words, output_dim=128, name="embedding_layer")(input_cnn)

# 1D Convolutional Layer with ReLU activation
conv = Conv1D(filters=128, kernel_size=5, activation='relu', name="conv1d_layer")(embedding)

# Global max pooling reduces sequence dimension to a fixed-size vector
pooling = GlobalMaxPooling1D(name="global_max_pooling")(conv)

# Dropout to prevent overfitting
drop1 = Dropout(0.5, name="dropout_1")(pooling)

# Fully connected layer with ReLU activation
dense = Dense(64, activation='relu', name="dense_relu")(drop1)

# Another dropout layer
drop2 = Dropout(0.5, name="dropout_2")(dense)

# Output layer with softmax for 3-class sentiment classification
output = Dense(3, activation='softmax', name="output_layer")(drop2)

# Create and compile the model
model_cnn = Model(inputs=input_cnn, outputs=output, name="CNN_Sentiment_Model")
model_cnn.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(learning_rate=1e-4),
    metrics=['accuracy']
)

# Model summary (optional)
model_cnn.summary()

# Train the model
model_cnn.fit(
    X_train_pad,
    y_train_cat,
    validation_data=(X_test_pad, y_test_cat),
    epochs=5,
    batch_size=32
)

pred_cnn = model_cnn.predict(X_test_pad)
pred_label_cnn = np.argmax(pred_cnn, axis=1)

print("\n[CNN Model Metrics]")
print(classification_report(true_labels, pred_label_cnn, target_names=label_encoder.classes_))
print("Accuracy:", accuracy_score(true_labels, pred_label_cnn))
print("F1 Score:", f1_score(true_labels, pred_label_cnn, average='weighted'))
print("Precision:", precision_score(true_labels, pred_label_cnn, average='weighted'))
print("Recall:", recall_score(true_labels, pred_label_cnn, average='weighted'))

sns.heatmap(confusion_matrix(true_labels, pred_label_cnn), annot=True, fmt='d', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("CNN Confusion Matrix")
plt.show()
```

```python
# ------------------------
# BERT QA Span Extraction
# ------------------------
class ReviewDataset(Dataset):
    def __init__(self, encodings):
        self.encodings = encodings
    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
    def __len__(self):
        return len(self.encodings['input_ids'])

questions = df['sentiment'].tolist()
contexts = df['text'].tolist()
answers = df['selected_text'].tolist()
start_positions = [context.find(answer) for context, answer in zip(contexts, answers)]
end_positions = [start + len(answer) for start, answer in zip(start_positions, answers)]

bert_tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
encodings = bert_tokenizer(questions, contexts, truncation=True, padding=True, return_offsets_mapping=True)
offsets = encodings.pop("offset_mapping")

start_pos, end_pos = [], []
for i, offset in enumerate(offsets):
    start, end = start_positions[i], end_positions[i]
    s_pos = e_pos = 0
    for idx, (s, e) in enumerate(offset):
        if s <= start < e:
            s_pos = idx
        if s < end <= e:
            e_pos = idx
            break
    start_pos.append(s_pos)
    end_pos.append(e_pos)

encodings.update({'start_positions': start_pos, 'end_positions': end_pos})
dataset = ReviewDataset(encodings)

model_qa = BertForQuestionAnswering.from_pretrained("bert-base-uncased")
training_args = TrainingArguments(output_dir="./results", num_train_epochs=1, per_device_train_batch_size=8,
                                  warmup_steps=10, weight_decay=0.01, logging_dir="./logs", logging_steps=10,
                                  save_strategy="no")

trainer = Trainer(model=model_qa, args=training_args, train_dataset=dataset)
trainer.train()
```
